В датафреймах присутствуют переменные с датами, которые желательно переконвертировать в тип данных datetime

Создам дополнительные столбцы с датами в формате год-месяц. Они понадобятся для группировки

```python
ads['ads_year_month'] = ads['created_at'].dt.strftime('%Y-%m')
leads['leads_year_month'] = leads['lead_created_at'].dt.strftime('%Y-%m')
purchases['purchases_year_month'] = purchases['purchase_created_at'].dt.strftime('%Y-%m')
```

```python
print(ads['ads_year_month'].head(1))
print(leads['leads_year_month'].head(1))
print(purchases['purchases_year_month'].head(1))
```

Для финальной таблички за основу возьмем датафрейм ads. На основе его меток и их содержимого будем джойнить с датафреймом leads.

```python
ads_clicks_cost = ads.groupby(['ads_year_month', 'd_utm_source', 'd_utm_medium', 'd_utm_campaign']).agg({'m_clicks': 'sum', 'm_cost': 'sum'}).reset_index().rename(columns={'m_clicks': 'clicks', 'm_cost': 'campaign_cost'})
ads_clicks_cost.head()
```

Проверим датафреймы на наличие дубликатов

```python
print(ads.duplicated().sum())
print(leads.duplicated().sum())
print(purchases.duplicated().sum())
```

Датафреймы ads и leads будем джойнить по следующим столбцам из ads(и аналогичным им из leads): created_at, d_utm_source, d_utm_medium, d_utm_campaign, d_utm_content<br>
Важно чтобы в обоих датафреймах значения этих столбцов совпадали

```python
print(ads['d_utm_source'].unique())
print(ads['d_utm_medium'].unique())
print(ads['d_utm_campaign'].nunique())
print(ads['d_utm_content'].nunique())
```

Перед тем соединить ads и leads, проведу чистку leads на предмет значений, отсутствующих в метках ads, а также NaN

```python
leads_cl = leads[(leads['d_lead_utm_source'] == 'yandex') & (leads['d_lead_utm_medium'] == 'cpc')]
leads_cl = leads_cl[leads_cl['d_lead_utm_campaign'].isin(ads['d_utm_campaign'].unique())]
leads_cl = leads_cl[leads_cl['d_lead_utm_campaign'].isin(ads['d_utm_campaign'].unique())]
leads_cl = leads_cl[leads_cl['d_lead_utm_content'].isin(ads['d_utm_content'].unique())]
```

```python
leads_cl.isnull().sum()
```

Удаляем столбец d_lead_utm_term из leads (в ads уже удалили - он был пустой). Также дропаем строки со значением NaN из client_id т.к. с отсутствующими данными не удастся сджойнить датафрейм с purchases по client_id

```python
leads_cl.drop('d_lead_utm_term', axis=1, inplace=True)
leads_cl.dropna(subset=['client_id'], axis='rows', inplace=True)
```

```python
leads_cl.isnull().sum()
```

Джойним датафремы ads и leads

```python
ads_leads = pd.merge(ads, leads_cl, how='inner', left_on=['created_at','d_utm_source', 'd_utm_medium', 'd_utm_campaign', 'd_utm_content'], right_on=['lead_created_at', 'd_lead_utm_source', 'd_lead_utm_medium', 'd_lead_utm_campaign', 'd_lead_utm_content'])
ads_leads.head()
```

```python
ads_leads.shape
```

Джойним ads_leads и purchases

```python
df = pd.merge(ads_leads, purchases, how='inner', on='client_id')
df.head()
```

```python
df.shape
```

Создаем столбец с кол-вом дней от создания лида до покупки

```python
df['lead_purchase_prd'] = (df['purchase_created_at'] - df['lead_created_at']).dt.days
df['lead_purchase_prd']
```

В переменной min_period держим purchase_id и минимальный период лида до покупки, чтобы соблюсти правило из задания

```python
min_period = df.groupby('purchase_id')['lead_purchase_prd'].min().reset_index().rename(columns={'lead_purchase_prd': 'min_lead_purchase_prd'})
```

```python
min_period
```

Джойним эту переменную с датафреймом

```python
df_2 = pd.merge(df, min_period, how='inner', on='purchase_id')
```

```python
df_2.head()
```

Атрибуция лид - покупка:<br>
**-Не должно быть отрицательных периодов**<br>
**-Минимальный период от лида до покупки**<br>
**-Период не больше 15 дней**<br>

```python
df_final = df_2[(df_2['lead_purchase_prd'] >= 0) & (df_2['lead_purchase_prd'] == df_2['min_lead_purchase_prd']) & (df_2['lead_purchase_prd'] <= 15)]
df_final.head(1)
```

```python
df_final.shape
```

Избавляемся от дубликатов

```python
df_final[df_final.duplicated(['client_id', 'lead_id'], keep=False)]
```

```python
df_final.drop_duplicates(['client_id', 'lead_id'], inplace=True)
```

```python
df_final.shape
```

```python
df_final['m_purchase_amount'].value_counts()
```

В столбце m_purchase_amount присутствуют покупки на нулевую сумму. От строк с такими данными стоит избавиться

```python
df_final = df_final[df_final['m_purchase_amount'] != 0]
```

```python
df_final.shape
```

Снова проверяем на дубликаты

```python
df_final[df_final.duplicated(['client_id', 'purchase_id'], keep=False)]
```

```python
df_final.drop_duplicates(['client_id', 'purchase_id'], inplace=True)
```

```python
df_final.shape
```

```python
df_final.sort_values( by='purchase_created_at', inplace=True)
```

```python
df_final.head()
```

Для финальной таблички нам важно знать кол-во лидов на рекламную кампанию

```python
num_leads = ads_leads.groupby(['ads_year_month', 'd_lead_utm_source', 'd_lead_utm_medium', 'd_lead_utm_campaign']).agg({'lead_id': 'nunique'}).reset_index().rename(columns={'lead_id': 'leads_count'})
num_leads.head(10)
```

Также нам важно знать кол-во уникальных purchase_id и сумму покупок

```python
sales = df_final.groupby(['ads_year_month', 'd_utm_source', 'd_utm_medium', 'd_utm_campaign']).agg({'purchase_id':'nunique', 'm_purchase_amount':'sum'}).reset_index().rename(columns={'purchase_id': 'sales_num', 
                                                                                                         'm_purchase_amount': 'revenue'})

sales.head(10)
```

Джойним ранее созданные группированные датафреймы

```python
df_grouped = pd.merge(ads_clicks_cost, sales, how='left', on=['ads_year_month', 'd_utm_source', 'd_utm_medium', 'd_utm_campaign'])
df_grouped.head(10)
```

```python
df_grouped = pd.merge(df_grouped, num_leads, how='left', left_on=['ads_year_month', 'd_utm_source', 
                                                                   'd_utm_medium', 'd_utm_campaign'], right_on=['ads_year_month', 'd_lead_utm_source', 
                                                                                                                                     'd_lead_utm_medium', 'd_lead_utm_campaign'])
```

```python
df_grouped.head(10)
```

Считаем CPL и ROAS

```python
df_grouped['CPL'] = (df_grouped['campaign_cost'] / df_grouped['leads_count']).round(2)
df_grouped['ROAS'] = (df_grouped['revenue'] / df_grouped['campaign_cost']).round(2)

df_grouped = df_grouped[['ads_year_month', 'd_utm_source', 'd_utm_medium', 'd_utm_campaign', 'clicks', 'campaign_cost','leads_count', 'revenue','CPL', 'ROAS']]
```

```python
df_grouped.head()
```

Убираем строки где стоимость компании равна нулю

```python
df_grouped = df_grouped[df_grouped['campaign_cost'] != 0]
```

```python
df_grouped
```
